```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
def plot_scores(fn):
    scores = np.load(fn)
    means = []
    for i in range(100,scores.shape[0]):
        means.append((i,np.mean(scores[i-100:i])))
    means = np.array(means)
    plt.figure(figsize=(10,6))
    plt.plot(scores,'.')
    plt.plot(means[:,0],means[:,1],'r')
    plt.grid()
    plt.xlabel('Episode')
    plt.ylabel('Score')
    plt.title(fn[:-4])
    plt.legend(['Episode Score', '100 Episode Averge Score'])
```

```python
plot_scores('visual_banana_scores.npy')
```

```python
plot_scores('basic_banana_scores.npy')
```
